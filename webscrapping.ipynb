```python
import requests
import re
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
```

```python
headers= {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0 Safari/537.36"
}
```

```python
Tv_name=[]
T_P=[]
rate=[]
cont=[]
```

```python
for i in range(1,21):
    url="https://www.flipkart.com/search?q=TV&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&p%5B%5D=facets.brand%255B%255D%3DLG&p%5B%5D=facets.brand%255B%255D%3DSONY&p%5B%5D=facets.brand%255B%255D%3DPanasonic&p%5B%5D=facets.brand%255B%255D%3DSamsung&p%5B%5D=facets.brand%255B%255D%3DXIAOMI&page="+str(i)
    req=requests.get(url,headers)
    content=BeautifulSoup(req.content,'html.parser')
    name=content.find_all('div',class_='KzDlHZ')
    price=content.find_all('div',class_='cN1yYO')
    ratings=content.find_all('div',class_='_5OesEi')
    con=content.find_all('div',class_='_6NESgJ')
  

    #appends
    for i in name:
        Tv_name.append(i.text)
    for i in price:
        T_P.append(i.text)
    for i in ratings:
        rate.append(i.text)
    for i in con:
        cont.append(i.text)
```

```python
Tv_name
```

```python
# TV NAME
```

```python
brand_name=[]
for i in Tv_name:
    regex=re.findall(r'^\w+',i)
    if regex:
        brand_name.append(regex[0])
    else:
       brand_name.append(np.nan)
len(brand_name)
```

```python
Inches=[]
for i in Tv_name:
    regex=re.findall(r'\((\d.+.nch)\)',i)
    if regex:
        Inches.append(regex[0])
    else:
       Inches.append(np.nan)
len(Inches)
```

```python
#Inches
```

```python
# PRICE
```

```python
c_p=[]
for i in T_P:
    regex=re.findall(r'.+₹(.+).\d% ',i)
    if regex:
        c_p.append(regex[0])
    else:
      c_p.append(np.nan)
len(c_p)
```

```python
s_p=[]
for i in T_P:
    regex=re.findall(r'₹(.+)₹',i)
    if regex:
        s_p.append(regex[0])
    else:
     s_p.append(np.nan)
len(s_p)
```

```python
Dis=[]
for i in T_P:
    regex=re.findall(r'.+₹.+(.\d)%',i)
    if regex:
        Dis.append(regex[0])
    else:
       Dis.append(np.nan)
len(Dis)
```

```python
led_model=[]
for i in cont :
    regex=re.findall(r'\| (.+) ID',i)
    if regex:
        led_model.append(regex[0])
    else:
       led_model.append(np.nan)
len(led_model)
            
```

```python
year=[]
for i in cont :
    regex=re.findall(r'Year: (\d{4})',i)
    if regex:
        year.append(regex[0])
    else:
       year.append(np.nan)
len(year)
```

```python
model_id=[]
for i in cont :
    regex=re.findall(r'ID: (.+)L',i)
    if regex:
        model_id.append(regex[0])
    else:
       model_id.append(np.nan)
len(model_id)
```

```python
r_data=pd.DataFrame({"Brand":brand_name,"Model ID":model_id,"LED Model":led_model,
                    "Inches":Inches,"YEAR":year,"Selling Price":s_p,"Original Cost":c_p,"Discount":Dis})
```
